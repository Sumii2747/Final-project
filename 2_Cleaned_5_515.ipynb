```python
import pandas as pd
import numpy as np
import Cleanup
```

```python
# define the function to be called later
clean = Cleanup.clean() #calling out the function 'Cleanup'
```

```python
df = pd.read_csv("EMD_Reconstructed(i80 0500-0515).csv") #read csv file
df.head()
```

```python
df = clean.filter_non_zero_values(df) # to remove or filter out any rows that contain non-zero values
```

```python
df
```

```python
df = clean.filter_by_desired_lanes(df) #keep only the rows in the DataFrame that correspond to the specified lane(s)
```

```python
df_filtered= clean.remove_lane_change(df) #df_filtered will contain the DataFrame with lane changes removed
```

```python
df_filtered
```

```python
l = ['Vehicle_ID', 'Frame_ID', 'Total_Frames', 'v_Class', 'Lane_ID', 'Preceding', 'Following']
for col in l:
    df_filtered[col] = df_filtered[col].astype(int) #converts all the values to integers
```

### Adding two new column (Location and Time period)


```python
num_rows = df_filtered.shape[0]
df_filtered["Location"] = ['i80']*num_rows #populates all rows in that column with the value 'i80 on all rows
df_filtered["Period"] = ['5:00PM - 5:15PM']* #num_rows populates all rows in that column with the value '5:00PM - 5:15PM' on all rows
```

```python
df_filtered.to_csv("Cleaned_5_515.csv") #saves the dataset into csv file named "Clean_5_515"
```
