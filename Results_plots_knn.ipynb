```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df_knn_5 = pd.read_csv("prediction_knn_5.csv")
df_knn_1 = pd.read_csv("prediction_knn_1.csv")
df_knn_15 = pd.read_csv("prediction_knn_15.csv")
df_knn_2 = pd.read_csv("prediction_knn_2.csv")
```

```python
# Step 1: Extract relevant columns
actual_acceleration = df_knn_5['v_Acc']
predicted_acceleration = df_knn_5['predicted_acceleration']

# Step 2: Filter data for reaction time of 0.5 seconds
reaction_time = 0.5  # Replace with the actual reaction time value in seconds
filtered_data = df_knn_5[df_knn_5['pair_Time_Duration'] == reaction_time]

# Extract filtered actual and predicted accelerations
filtered_actual_acceleration = filtered_data['v_Acc']
filtered_predicted_acceleration = filtered_data['predicted_acceleration']

# Step 3: Plot the actual and predicted accelerations
plt.figure(figsize=(10, 6))
plt.plot(filtered_actual_acceleration, label='Actual Acceleration', color="red")
plt.plot(filtered_predicted_acceleration, label='Predicted Acceleration', color="blue")
plt.xlabel('Time Frames')
plt.ylabel('Acceleration')
plt.title(f'Actual vs. Predicted Acceleration for Reaction Time: {reaction_time} seconds (KNN)')
plt.legend()
plt.show()

# Plots for Velocity for reaction time 0.5 seconds
actual_acceleration_1 = df_knn_5['v_Vel']
predicted_acceleration_1 = df_knn_5['predicted_velocity']

# Step 2: Filter data for reaction time of 0.5 seconds
reaction_time = 0.5  # Replace with the actual reaction time value in seconds
filtered_data_1 = df_knn_5[df_knn_5['pair_Time_Duration'] == reaction_time]

# Extract filtered actual and predicted accelerations
filtered_actual_acceleration_1 = filtered_data_1['v_Vel']
filtered_predicted_acceleration_1 = filtered_data_1['predicted_velocity']

# Step 3: Plot the actual and predicted accelerations
plt.figure(figsize=(10, 6))
plt.plot(filtered_actual_acceleration_1, label='Actual Velocity', color="red")
plt.plot(filtered_predicted_acceleration_1, label='Predicted Velocity', color="blue")
plt.xlabel('Time Frames')
plt.ylabel('Actual Velocity')
plt.title(f'Actual vs. Predicted Velocity for Reaction Time: {reaction_time} seconds (KNN)')
plt.legend()
plt.show()

# Plot for Spacing for reaction time 0.5 seconds
actual_acceleration_2 = df_knn_5['nextFrameSpacing']
predicted_acceleration_2 = df_knn_5['predicted_spacing']

# Step 2: Filter data for reaction time of 0.5 seconds
reaction_time = 0.5  # Replace with the actual reaction time value in seconds
filtered_data_2 = df_knn_5[df_knn_5['pair_Time_Duration'] == reaction_time]

# Extract filtered actual and predicted accelerations
filtered_actual_acceleration_2 = filtered_data_2['nextFrameSpacing']
filtered_predicted_acceleration_2 = filtered_data_2['predicted_spacing']

# Step 3: Plot the actual and predicted accelerations
plt.figure(figsize=(10, 6))
plt.plot(filtered_actual_acceleration_2, label='Actual Spacing', color="red")
plt.plot(filtered_predicted_acceleration_2, label='Predicted Spacing', color="blue")
plt.xlabel('Time Frames')
plt.ylabel('Actual Spacing')
plt.title(f'Actual vs. Predicted Spacing for Reaction Time: {reaction_time} seconds (KNN)')
plt.legend()
plt.show()
```
